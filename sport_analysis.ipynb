```python
prova = sport_dataset.groupby(["id", "first2w"], as_index= False).count()
```

```python
prova
# we are able to see how many surveys the user has answered to for each week claiming that he's been doing sport-
# if we have data for both weeks we consider the user as habitual, otherwise as occasional
```

```python
first_week = prova[prova.first2w == "First two weeks"]
second_week = prova[prova.first2w == "Second two weeks"]

user_ids = list(prova["id"].unique())
count_first_week = list(first_week["pilot"].values)
count_second_week = list(second_week["pilot"].values)
user_classification = []
sport_hours = []

for i in range(len(user_ids)):
    current_id = user_ids[i]
    sport_first = count_first_week[i]
    sport_second = count_second_week[i]
    
    if sport_first!= 0 and sport_second!=0:
        user_classification.append("habitual")
    else:
        user_classification.append("occasional")
    
    current_hours = (sport_first + sport_second)/2
    sport_hours.append(current_hours)
        

str_id = [str(int(el)) for el in user_ids]

cleaned_dataset = {"id": pd.Series(user_ids),
                   "userid": pd.Series(str_id),
                   "first week": pd.Series(count_first_week),
                   "second week": pd.Series(count_second_week),
                   "classification": pd.Series(user_classification),
                   "hours float": pd.Series(sport_hours),
                   "hours int": pd.Series([int(el) for el in sport_hours])}

final_dataset = pd.DataFrame(data=cleaned_dataset)
```

```python
final_dataset
```

```python
sportive_ids = list(final_dataset["userid"].values)
sportive_categories = list(final_dataset["classification"].values)
sportive_categories_ids = {}

index = 0
for i in sportive_ids:
    #current = {}
    
    sportive_categories_ids[i] = sportive_categories[index]
    index+=1
    #sportive_categories_ids.append(current)
    
total_ids = list(demo_dataset["userid"].values)
```

```python
non_sportive_ids = set(total_ids) - set(sportive_ids)
```

```python
for i in non_sportive_ids:
    sportive_ids.append(i)
    sportive_categories.append("non sportive")
```

```python
sportive_categories_num, uniques = pd.factorize(pd.Series(sportive_categories), sort=True)


sportive_data = {"userid": pd.Series(sportive_ids),
                "sport cat": pd.Series(sportive_categories),
                "sport num": pd.Series(sportive_categories_num)}

sportive_dataset = pd.DataFrame(data=sportive_data)
```
